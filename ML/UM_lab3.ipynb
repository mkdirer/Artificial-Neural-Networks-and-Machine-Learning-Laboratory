```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt

df = pd.DataFrame()
df['x']=[1,2,3,4,7,8,10]
df['y']=[4,6,7,10,13,18,20] 
```

```python
plt.scatter(df['x'], df['y'], label = 'Wartosc')
plt.xlabel('Wartosc X')
plt.xlabel('Wartosc Y')
plt.legend()
plt.show()
#plt.plot(df['x'],df['y'], '.')
```

```python
mean_x = np.mean(df['x'])
```

```python
mean_y = np.mean(df['y'])
```

```python
S_x = np.std(df['x'])
S_y = np.std(df['y'])
print(S_y)
print(S_x)
```

```python
r = np.corrcoef(df['x'], df['y'])
r = df['x'].corr(df['y'])
```

```python
b = (S_y/S_x) * r
a = mean_y - b * mean_x
```

```python
def linregression(a,b,x):
  return a + b * x
```

```python
x = np.linspace(0, 10, 100)


plt.plot(x, linregression(a, b, x), color ='g')
plt.scatter(df['x'], df['y'], label = "wartosc")
```

```python
df.append({'x': 15, 'y': np.nan}, ignore_index=True)
```

```python
df.at[10, 'x'] = 22
df.at[10, 'y'] = linregression(a, b, df['x'][10])
 
print(df)
```

```python
x = np.linspace(0, 10, 100)


plt.plot(x, linregression(a, b, x), color ='g')
plt.scatter(df['x'], df['y'], label = "wartosc")
```

```python
def calculate_r_squere(a, b, df, mean_y):
  SSm = 0.0
  SSt = 0.0
  for _, row in df.iterrows():
    x = row['x']
    y = row['y']
    y_p = linregression(a,b, x)
    SSm += (y_p - mean_y) ** 2
    SSt += (y - mean_y) ** 2
  return SSm / SSt

r2 = calculate_r_squere(a,b,df,mean_y)
print("r2", r2)

print("wspolcznnik zbieznosci: ", 1 - r2)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(df["y"], linregression(a, b, df['x']))
print(r2)
```
