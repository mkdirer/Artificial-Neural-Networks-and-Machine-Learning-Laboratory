```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans

digits = load_digits()
X = digits.data
y = digits.target

plt.gray()
plt.matshow(digits.images[30])
plt.show()

kmeans = KMeans(n_clusters=10)
clusters = kmeans.fit_predict(X)

fig, ax = plt.subplots(2, 5, figsize=(10, 4))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)

for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='bicubic', cmap=plt.cm.gray_r)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from skfuzzy.cluster import cmeans
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target

plt.gray()
plt.matshow(digits.images[30])
plt.show()

n_clusters = 10
cntr, u, u0, d, jm, p, fpc = cmeans(X.T, n_clusters, 2, error=0.005, maxiter=1000)

fig, ax = plt.subplots(2, 5, figsize=(10, 4))
centers = cntr.T.reshape(n_clusters, 8, 8)

for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='bicubic', cmap=plt.cm.gray_r)

plt.tight_layout()

fig2, ax2 = plt.subplots()
ax2.set_title('Fuzzy Membership Matrix')
ax2.set_xlabel('Cluster')
ax2.set_ylabel('Sample')
im2 = ax2.imshow(u, cmap='viridis', aspect='auto')
cbar2 = plt.colorbar(im2, ax=ax2, fraction=0.046, pad=0.04)
cbar2.set_label('Membership Degree')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.cluster import AgglomerativeClustering
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target

n_clusters = 10
clustering = AgglomerativeClustering(n_clusters=n_clusters)
clusters = clustering.fit_predict(X)

fig, ax = plt.subplots(2, 5, figsize=(10, 4))
centers = []
data_counts = np.zeros(n_clusters)  # Licznik danych w poszczeg√≥lnych klastrach

for i in range(n_clusters):
    cluster_points = X[clusters == i]
    data_counts[i] = len(cluster_points)  # Zliczanie liczby danych w klastrze
    center = cluster_points.mean(axis=0)
    centers.append(center)
    axi = ax[i // 5, i % 5]
    axi.set(xticks=[], yticks=[])
    axi.imshow(center.reshape(8, 8), interpolation='bicubic', cmap=plt.cm.gray_r)

# Wy≈õwietlanie liczby danych na wykresie
for i, axi in enumerate(ax.flat):
    axi.text(0.05, 0.05, f'N={int(data_counts[i])}', transform=axi.transAxes, color='red')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.mixture import GaussianMixture
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target

n_clusters = 10
gmm = GaussianMixture(n_components=n_clusters)
gmm.fit(X)
clusters = gmm.predict(X)

fig, ax = plt.subplots(2, 5, figsize=(10, 4))
centers = gmm.means_

data_counts = np.zeros(n_clusters)  # Licznik danych w poszczeg√≥lnych klastrach

for i in range(n_clusters):
    cluster_points = X[clusters == i]
    data_counts[i] = len(cluster_points)  # Zliczanie liczby danych w klastrze
    center = centers[i]
    axi = ax[i // 5, i % 5]
    axi.set(xticks=[], yticks=[])
    axi.imshow(center.reshape(8, 8), interpolation='bicubic', cmap=plt.cm.gray_r)

# Wy≈õwietlanie liczby danych na wykresie
for i, axi in enumerate(ax.flat):
    axi.text(0.05, 0.05, f'N={int(data_counts[i])}', transform=axi.transAxes, color='red')

plt.tight_layout()
plt.show()
```
