```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# a = np.array([[0.0, 0.0],
#               [3.0, 4]])
# b = np.array([[1.1, 1.0],
#               [4.0, 4.5]])

dot_ab = a @ b
print(dot_ab)
#dot_ab = np.dot(a, b)
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
sin_ab = dot_ab / (norm_a * norm_b)
angle_ab = np.arcsin(sin_ab)

print("Sinus kƒÖta miƒôdzy wektorami a i b wynosi:", sin_ab)
print("KƒÖt miƒôdzy wektorami a i b wynosi:", angle_ab)
angle_ab_degrees = np.degrees(angle_ab)
print("KƒÖt miƒôdzy wektorami a i b wynosi:", angle_ab_degrees, "¬∞")
```
