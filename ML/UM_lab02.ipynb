```python
import numpy as np
import matplotlib.pyplot as plt

# Tworzenie macierzy 100 z losowymi wartościami zmiennoprzecinkowymi
matrix = np.random.normal(0, 1, 100)

# Wyznaczenie średniej, wariancji i wartości maksymalnej
mean = np.mean(matrix)
variance = np.var(matrix)
maximum = np.max(matrix)

# Wyświetlenie wyników
print("Średnia: ", mean)
print("Wariancja: ", variance)
print("Wartość maksymalna: ", maximum)

# Histogram
plt.hist(matrix.flatten(), bins=20)
plt.title("Histogram wartości macierzy")
plt.xlabel("Wartość")
plt.ylabel("Liczba wystąpień")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tworzenie tablicy jednowymiarowej o 100 elementach z losowymi wartościami zmiennoprzecinkowymi
arr = np.random.random(10)
arr1 = np.random.normal(0, 5,10)
arr2 = np.random.normal(0, 5,10)
arr3 = np.random.normal(0, 5,10)
x = np.arange(10)

yerr = np.linspace(0.05, 0.2, 10)

plt.errorbar(x, arr, yerr=yerr, label='Oryginal')

plt.errorbar(x, arr+ 1, yerr=yerr, uplims=True, label='przesuniecie o +1')

plt.errorbar(x, arr+2, yerr=yerr, uplims=True, lolims=True,
             label='przesuniecie o +2')

upperlimits = [True, False] * 5
lowerlimits = [False, True] * 5
plt.errorbar(x, arr+3, yerr=yerr, uplims=upperlimits, lolims=lowerlimits,
             label='przesuniecie o +3')

plt.legend(loc='lower right')
```

```python
import matplotlib.pyplot as plt

# Dane
data = [('2021-10-01', 1.2), ('2021-11-04', 1.3), ('2021-12-07', 2.1), ('2022-02-01', 0.4), ('2022-03-15', 3.1)]

# Konwersja dat do indeksów numerycznych
x = range(len(data))
# Wyciągnięcie wartości Y
y = [d[1] for d in data]
# Wyciągnięcie etykiet X
labels = [d[0] for d in data]

# Tworzenie wykresu
fig, ax = plt.subplots()
ax.plot(x, y, color='blue', linestyle='solid', marker='o', markersize=12)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.set_xlabel('Data')
ax.set_ylabel('Wartość')
ax.set_title('Wykres punktowy')
ax.grid(axis='both')

plt.show()
```

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset("tips")

# Wartości mediany, średniej, minimalnej i maksymalnej dla zmiennej total_bill
median_total_bill = tips["total_bill"].median()
mean_total_bill = tips["total_bill"].mean()
min_total_bill = tips["total_bill"].min()
max_total_bill = tips["total_bill"].max()

print(f"Median: {median_total_bill}")
print(f"Mean: {mean_total_bill}")
print(f"Minimum: {min_total_bill}")
print(f"Maximum: {max_total_bill}")
```

```python
# Wartości mediany, średniej, minimalnej i maksymalnej dla zmiennej total_bill dla kobiet i mężczyzn
median_total_bill_female = tips.loc[tips["sex"] == "Female", "total_bill"].median()
mean_total_bill_female = tips.loc[tips["sex"] == "Female", "total_bill"].mean()
min_total_bill_female = tips.loc[tips["sex"] == "Female", "total_bill"].min()
max_total_bill_female = tips.loc[tips["sex"] == "Female", "total_bill"].max()

median_total_bill_male = tips.loc[tips["sex"] == "Male", "total_bill"].median()
mean_total_bill_male = tips.loc[tips["sex"] == "Male", "total_bill"].mean()
min_total_bill_male = tips.loc[tips["sex"] == "Male", "total_bill"].min()
max_total_bill_male = tips.loc[tips["sex"] == "Male", "total_bill"].max()

print("Female:")
print(f"Median: {median_total_bill_female}")
print(f"Mean: {mean_total_bill_female}")
print(f"Minimum: {min_total_bill_female}")
print(f"Maximum: {max_total_bill_female}")

print("\nMale:")
print(f"Median: {median_total_bill_male}")
print(f"Mean: {mean_total_bill_male}")
print(f"Minimum: {min_total_bill_male}")
print(f"Maximum: {max_total_bill_male}")
```
